```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.fujisafari.co.jp/'
response = requests.get(url)
response.raise_for_status()

soup = BeautifulSoup(response.content, 'html.parser')

info_section_inner = soup.find('div', class_='info-section__inner')
info_schedule_elements = info_section_inner.find_all('dl', class_='info-schedule')
print(f'info_schedule_elements{info_schedule_elements}')

data_list = []

#dictionary data(dd,dt)
data_dic={}
for tag in info_section_inner.find_all(['dt','dd']):
        if tag.name == 'dt':
                key = tag.get_text()
        if tag.name == 'dd':
                data_dic[key] = tag.get_text().strip()
print('----------------------------------------')
print(data_dic)

#これで、ddに対するdtのコンテンツを表示。この場合は改行,スペースを考慮して取得すればよい
print(data_dic['\n\n                    8月15日（火）\n                \n'])
print('----------------------------------------')


'''chatGPT'''
for info_schedule in info_schedule_elements:
    date_element = info_schedule.find('dt', class_='info-schedule__title')
    date = date_element.span.text.strip()

    reception_time_element = info_schedule.find('h3', string='入園受付時間')
    if reception_time_element:
        reception_time = reception_time_element.find_next('div', class_='info-schedule__time__wrap').text.strip()
    else:
        reception_time = "情報なし"

    temperature_element = info_schedule.find('h3', class_='info-schedule__time__title')
    if temperature_element:
        temperature_time = temperature_element.get_text().strip()
        temperature = temperature_element.find_next('p', class_='temperature').text.strip()
    else:
        temperature_time = "情報なし"
        temperature = "情報なし"

    data = {
        'date': date,
        'reception_time': reception_time,
        'temperature_time': temperature_time,
        'temperature': temperature
    }
    data_list.append(data)

# データをプリント
for data in data_list:
    print("日付:", data['date'])
    print("入園受付時間:", data['reception_time'])
    print("気温時刻:", data['temperature_time'])
    print("気温:", data['temperature'])
    print("-" * 20)
```

```python
#chatGPT結果
import requests
from bs4 import BeautifulSoup

url = 'https://www.fujisafari.co.jp/'
response = requests.get(url)
response.raise_for_status()

soup = BeautifulSoup(response.content, 'html.parser')

info_section_inner = soup.select_one('div.info-section__inner')
info_schedule_elements = info_section_inner.select('dl.info-schedule')

data_list = []

for info_schedule in info_schedule_elements:
    date_element = info_schedule.select_one('dt.info-schedule__title span')
    date = date_element.get_text(strip=True)

    reception_time_element = info_schedule.select_one('div.info-schedule__time__wrap')
    if reception_time_element:
        reception_time = reception_time_element.get_text(strip=True)
    else:
        reception_time = "情報なし"

    temperature_element = info_schedule.select_one('h3.info-schedule__time__title:contains("気温")')
    if temperature_element:
        temperature_time = temperature_element.get_text(strip=True)
        temperature = temperature_element.find_next('p', class_='temperature').get_text(strip=True)
    else:
        temperature_time = "情報なし"
        temperature = "情報なし"

    data = {
        'date': date,
        'reception_time': reception_time,
        'temperature_time': temperature_time,
        'temperature': temperature
    }
    data_list.append(data)

# データをプリント
for data in data_list:
    print("日付:", data['date'])
    print("入園受付時間:", data['reception_time'])
    print("気温時刻:", data['temperature_time'])
    print("気温:", data['temperature'])
    print("-" * 20)
```
