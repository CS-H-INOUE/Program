```python
#chatGPTから一部改修
#find ->　select(css selector使用)

import requests
from bs4 import BeautifulSoup

url = 'https://www.fujisafari.co.jp/'
response = requests.get(url)
response.raise_for_status()

soup = BeautifulSoup(response.content, 'html.parser')

info_section_inner = soup.select_one('div.info-section__inner')
info_schedule_elements = info_section_inner.select('dl.info-schedule')

data_list = []

for info_schedule in info_schedule_elements:
    date_element = info_schedule.select_one('dt.info-schedule__title span')
    date = date_element.get_text(strip=True)

    reception_time_element = info_schedule.select_one('div.info-schedule__time__wrap')
    if reception_time_element:
        reception_time = reception_time_element.get_text(strip=True)
    else:
        reception_time = "情報なし"

    temperature_element = info_schedule.select_one('h3.info-schedule__time__title:contains("気温")')
    if temperature_element:
        temperature_time = temperature_element.get_text(strip=True)
        temperature = temperature_element.find_next('p', class_='temperature').get_text(strip=True)
    else:
        temperature_time = "情報なし"
        temperature = "情報なし"

    data = {
        'date': date,
        'reception_time': reception_time,
        'temperature_time': temperature_time,
        'temperature': temperature
    }
    data_list.append(data)

# データをプリント
for data in data_list:
    print("日付:", data['date'])
    print("入園受付時間:", data['reception_time'])
    print("気温時刻:", data['temperature_time'])
    print("気温:", data['temperature'])
    print("-" * 20)
```

```python
######ここから、複数データのスクレイピング#####

import requests
from bs4 import BeautifulSoup

res = requests.get('https://book.impress.co.jp/booklist/')
html_doc =res.text
soup = BeautifulSoup(html_doc, 'html.parser')
div_book_list = soup.find('div', class_= 'block-book-list-body')

book_urls = []
a_tags = div_book_list.find_all('a')
for a_tag in a_tags:
    if a_tag['href'] not in book_urls:
        book_urls.append(a_tag['href'])

```

```python
def get_book_info(book_url):
    res=requests.get(book_url)
    html_doc = res.text
    soup = BeautifulSoup(html_doc,'html.parser')
    
    div_book_detail = soup.find('div', class_='block-book-detail')
    book_title = div_book_detail.find('h2')
    book_price = div_book_detail.find('p', class_= 'module-book-price')
    
    book_data = {}
    dl_book_data = div_book_detail.find('dl', class_='module-book-data')
    for tag in dl_book_data.find_all(['dt', 'dd']):
        if tag.name == 'dt':
            key = tag.get_text()
        if tag.name == 'dd':
            book_data[key] = tag.get_text().strip()
    
    return [
        book_title.get_text(),
        book_price.get_text(),
        book_data['発売日'],
        book_data['著者']
    ]
```

```python
get_book_info('https://book.impress.co.jp/books/1123101013')
```

```python
import time

book_info_list =[]
for book_url in book_urls:
    print(f'Scraping NOW!!!!!: {book_url}')
    book_info_list.append(get_book_info(book_url))
    time.sleep(1)
print('finished!')

```

```python
#\t: tabでjoin 1collumn 1data -> \nで制御
with open('book_data.tsv', 'w', encoding='utf-8') as f:
    for book_info in book_info_list:
        f.write('\t'.join(book_info) + '\n') 
```

```python
def book_search(keyword):
    results =[]
    with open('book_data.tsv', encoding='utf-8')as f:
        for line in f:
            cols = line.split('\t')
            if keyword in cols[0]:
                results.append(line)
    if len(results) > 0:
        response = ''.join(results)
    else:
        response = '「{}」はnothing'
    return response
```
